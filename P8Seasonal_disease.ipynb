```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Create simple data
data = {
    'Temp': [25,28,30,32,35,33,29,27,26,31],
    'Rainfall': [10,20,50,200,300,250,100,30,20,150],
    'Cases': ['Low','Low','Medium','High','High','High','Medium','Low','Low','Medium']
}

df = pd.DataFrame(data)

# Convert text to numbers
df['Cases'] = df['Cases'].map({'Low':0,'Medium':1,'High':2})

# Train model
model = RandomForestClassifier()
model.fit(df[['Temp','Rainfall']], df['Cases'])

print("Model Ready!")
```

```python
temp = float(input("Enter Temperature: "))
rain = float(input("Enter Rainfall: "))

pred = model.predict([[temp, rain]])

labels = ['Low','Medium','High']
print("Disease Risk:", labels[pred[0]])
```
